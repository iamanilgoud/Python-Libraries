## 3 Bar Chart

##### Simple bar chart showing carb and gear.

```python
cars.head()
```

```python
model =cars.model
mileage =cars.mpg


plt.bar(model,mileage,label="cars_data")

plt.title('Model vs MPG')
plt.xlabel('Number of cars Models')
plt.ylabel( 'MPG')
plt.legend()


```

## 4 Histogram

##### In histograms, x axis contains a variable and y axis will be a frequency of that variable

```python
model= (cars.model)


plt.hist(model,rwidth=0.5)
plt.title('Histogram for Car models')
plt.xlabel('Model of the cars')
plt.ylabel('No.of cars')
plt.show()
```

## 5 Matplotlib Tutorial: Pie Charts

```python
horse_power =cars.hp
model = cars.model
plt.pie(horse_power,labels=model)
```

```python
#Draw pie chart as perfect circle
plt.pie(horse_power,labels=model, shadow=True)
plt.axis("equal")
plt.show()
```

```python
#Show percentages for every pie. Specify radius to increase chart sizet
plt.axis("equal")
plt.pie(horse_power,labels=model, shadow=True, autopct='%1.1f%%',radius=1.5)
plt.show()
```

## 6 Area chart 

##### . Plot the area chart for the colmns: am and carb.
a. Set the am on the x-axis. 
b. Set the carb on the y-axis. 
c. Provide the x-axis label as Transmission. 
d. Provide the y-axis labe as Number of carburetors. 
e. Provide the title as Transmission vs Number of carburetors.

```python
am=(cars.am)
carb= (cars.carb)

plt.stackplot(am,carb)
plt.title('Transmission vs Number of carburetors')
plt.xlabel('Transmission')
plt.ylabel('Number of carburetors')
plt.show()

plt.stackplot(am,carb)
plt.savefig(r"C:\Users\HP\Desktop\python\Matplotlib/areachart.jpg", bbox_inches="tight", pad_inches=10, transparent=True)
```
